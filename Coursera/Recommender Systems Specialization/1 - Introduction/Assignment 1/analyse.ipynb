```python
import pandas as pd
```

```python
df = pd.read_csv("HW1-data.csv")
```

```python
df_clean = df.melt(id_vars=["User", "Gender (1 =F, 0=M)"], var_name="movie", value_name="rating")
```

```python
df_clean.groupby("movie")["rating"].mean().round(3).reset_index().sort_values("rating")
```

```python
df_clean.groupby("movie")["rating"].count().round(3).reset_index().sort_values("rating")
```

```python
df_clean["good"] = df_clean["rating"] >= 4
```

```python
df_clean[df_clean["rating"].notnull()].groupby("movie")["good"].mean().sort_values()
```

```python
P_toy = df[
    (df["1: Toy Story (1995)"].notnull())
    ].shape[0] / df["User"].nunique()
```

```python
res = {}
for movie in df_clean["movie"].unique():
    P_movie_toy = df[
        (df["1: Toy Story (1995)"].notnull()) &
        (df[movie].notnull())
        ].shape[0] / df["User"].nunique()
    res[movie] = P_movie_toy / P_toy
```

```python
pd.DataFrame(list(res.items()), columns=['movie', 'score']).sort_values("score")
```
